```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import numpy as np
```

```python
df_energy = pd.read_parquet(f'app\Data\energy_demand.parquet',  engine='pyarrow')
```

```python
df_energy = df_energy.drop('respondent-name', axis=1)
```

```python
df_ny = df_energy.query("respondent == 'NYIS'")
df_ca = df_energy.query("respondent == 'CISO'")
df_tx = df_energy.query("respondent == 'ERCO'")
df_fl = df_energy.query("respondent == 'FLA'")
```

```python
winter_tx_2021 = df_tx.loc[(df_tx['period'] > '2021-02-08') & (df_tx['period'] < '2021-02-27')] 
```

```python
fig = px.line(winter_tx_2021, x='period', y='value')
fig.add_hline(y=winter_tx_2021['value'].max(), line_dash='dot', annotation_text = f'peak power load: {round(winter_tx_2021.value.max())} megawatthours', annotation_position='top')
fig.add_vrect(x0='2021-02-15T05', x1='2021-02-18', annotation_text='rolling outages', annotation_position='bottom right', fillcolor='green', opacity=0.20, line_width=0, )
fig.update_layout(title='Winter Storm of Texas 2021', xaxis_title='Date', yaxis_title='megawatthours',)
fig.show()
```

```python
summer_ca_2022 = df_ca.loc[(df_ca['period'] > '2022-08-24') & (df_ca['period'] < '2022-09-15')]
```

```python
fig = px.line(summer_ca_2022, x='period', y='value')
fig.add_hline(y=summer_ca_2022['value'].max(), 
              line_dash='dot', 
              annotation_text = f'peak power load: {summer_ca_2022.value.max()} megawatthours')
fig.add_vline(x='2022-09-07T02', 
              line_dash="dash", 
              # annotation_text = 'text', 
              line_color="red")
fig.add_annotation(x='2022-09-07T02', y=51_000,
            text="Emergency Text sent, Sept 6, 2023 @ 1745 Local Time",
            showarrow=True
            )
fig.add_vrect(x0='2022-08-30', x1='2022-09-09', annotation_text='Heatwave, Aug 30, 2022 - Sept 9, 2022', annotation_position='bottom left', fillcolor='orange', opacity=0.20, line_width=0, )
fig.update_layout(title='Power Disruption Prevention in California', xaxis_title='Date', yaxis_title='megawatthours',)
fig.show()
```
